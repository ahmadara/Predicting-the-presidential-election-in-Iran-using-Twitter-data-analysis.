```python
final_raisi = pd.read_csv('final_raisi.csv') 
final_hemati = pd.read_csv('final_hemati.csv')
final_mehr = pd.read_csv('final_mehr.csv')
final_rezai = pd.read_csv('final_rezai.csv')
final_jalili = pd.read_csv('final_jalili.csv')
final_zakani = pd.read_csv('final_zakani.csv')
final_hashemi = pd.read_csv('final_hashemi.csv')
```

```python
model_names = "persiannlp/mt5-large-parsinlu-sentiment-analysis"
tokenizers = MT5Tokenizer.from_pretrained(model_names)
models = MT5ForConditionalGeneration.from_pretrained(model_names)
```

```python
def run_model(context, **generator_args):
    input_ids = tokenizers.encode(context , return_tensors="pt")
    res = models.generate(input_ids, **generator_args)
    output = tokenizers.batch_decode(res, skip_special_tokens=True)
    return output
```

```python
unid = list(final_raisi['ID'])
```

```python
def vote_score(res):
  score = 0
  for text in res:
    # print(text)
    if len(text) > 3:
      text = text.replace("'", "")
      sen = run_model(text)
      sentiment = sen[0]
      if sentiment == 'very negative':
        score = score + -2
      elif sentiment == 'negative':
        score = score + -1
      elif sentiment == 'neutral':
        score = score + 0
      elif sentiment == 'mixed':
        score = score +  0
      elif sentiment == 'positive':
        score = score + 1
      elif sentiment == 'very positive':
        score = score + 2
      elif sentiment == 'no sentiment expressed':
        score = score + 0.1
      # print(sentiment)
      # print('*************')
  return score
```

```python
start_id = 7000
end_id = 7339
name = 'saved8.csv'

raisi_vote = 0
hemati_vote = 0
mehr_vote = 0
rezai_vote = 0
jalili_vote = 0
zakani_vote = 0
hashemi_vote = 0
vote = []
all = pd.DataFrame()

for i in range(start_id,end_id):
  print('\n')
  print(i)
  res = final_raisi['Raisi Tweets'][i].strip('][').split(', ')
  if res:
    # print('raisiiiiiii')
    raisi_vote = vote_score(res)

  res = final_hemati['Hemati Tweets'][i].strip('][').split(', ')
  if res:
    # print('hematiiiiii')
    hemati_vote = vote_score(res)

  res = final_mehr['Mehr Tweets'][i].strip('][').split(', ')
  if res:
    # print('meehrrr')
    mehr_vote = vote_score(res)

  res = final_rezai['Rezai Tweets'][i].strip('][').split(', ')
  if res:
    # print('rezaiiiii')
    rezai_vote = vote_score(res)

  res = final_jalili['Jalili Tweets'][i].strip('][').split(', ')
  if res:
    # print('jaliliiii')
    jalili_vote = vote_score(res)

  res = final_zakani['Zakani Tweets'][i].strip('][').split(', ')
  if res:
    # print('zakaniiii')
    zakani_vote = vote_score(res)

  res = final_hashemi['Hashemi Tweets'][i].strip('][').split(', ')
  if res:
    # print('hashemiiii')
    hashemi_vote = vote_score(res)

  noms = np.array([raisi_vote, hemati_vote, mehr_vote, rezai_vote, jalili_vote, zakani_vote, hashemi_vote])
  max_ind = np.argmax(noms)
  if noms[max_ind] == 0:
    vote.append(100)
  else:
    vote.append(max_ind)

all = pd.concat([final_raisi[start_id:end_id], final_hemati['Hemati Tweets'][start_id:end_id], final_mehr['Mehr Tweets'][start_id:end_id], final_rezai['Rezai Tweets'][start_id:end_id], final_jalili['Jalili Tweets'][start_id:end_id], final_zakani['Zakani Tweets'][start_id:end_id], final_hashemi['Hashemi Tweets'][start_id:end_id]], axis = 1)
all['vote'] = vote
all.to_csv('Saved/' + name, index = False)
```
