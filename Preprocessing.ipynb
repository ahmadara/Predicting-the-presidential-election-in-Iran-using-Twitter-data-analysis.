# Remove RT, Comments, and None

```python
def delrt(sentences):
  indexes = []
  for i in range(len(sentences)):
    st = sentences['text'][i]
    try:
      if st.startswith('RT') or st != st or st.startswith('@'):
        indexes.append(i)
    except:
      indexes.append(i)

  sentences = sentences.drop(indexes)
  sentences = sentences.reset_index(drop=True)
  return sentences
```

```python
hashemi = delrt(hashemi)
rezai = delrt(rezai)
mehr = delrt(mehr)
jalili = delrt(jalili)
raisi = delrt(raisi)
zakani = delrt(zakani)
hemati = delrt(hemati)
```

# Remove other languages

```python
def delan(sentences):
  indexes = []
  for i in range(len(sentences)):
    if sentences['lang'][i] != 'fa':
      indexes.append(i)

  sentences = sentences.drop(indexes)
  sentences = sentences.reset_index(drop=True)
  return sentences
```

```python
hashemi = delan(hashemi)
rezai = delan(rezai)
mehr = delan(mehr)
jalili = delan(jalili)
raisi = delan(raisi)
zakani = delan(zakani)
hemati = delan(hemati)
```

```python
from hazm import word_tokenize, stopwords_list, InformalLemmatizer
import re
#import arabic_reshaper
from tqdm import tqdm as progressbar
from hazm import Normalizer

my_file = open("stopwords.txt", "r", encoding='utf-8')
content_list = my_file.readlines()
sw=[]
for w in content_list:
    sw.append(w.strip())
sw.append('ها')

def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)


def remove_hashtag(text):
    return re.sub(r'#([^\s]+)', r'\1', text)

def remove_extra_repeated_alpha(text):
    return re.sub(r'([^\W\d_])\1{2,}', r'\1', text)

def remove_punctuations(text):
    return re.sub(r'[^\w\s]', ' ', text)

def remove_latin(text):
    output = re.sub(r'[A-Za-z]+',' ', text)
    output = re.sub(r'[0-9]+',' ', output)
    #output = re.sub(r"^[ آابپتثجچحخدذرزژسشصضطظعغفقکگلمنوهی]+$", "", text)
    output.rstrip()
    return output

lemma = InformalLemmatizer()
normalizer = Normalizer()

def clean_up(text,norm =False, latin = True, emoji = False, url=True, html=True, punctuations=True, lemmatize=False, hashtag=True, stopwords=False, remove_extra_alpha=True):
    # remove emoji
    if emoji:
        text = deEmojify(text)
    # remove url
    if url:
        text = remove_urls(text)
    # remove html tags
    if html:
        text = remove_html(text)
        
    text = text.replace("_", " ")
    text = text.replace("-", " ")
    # remove punctuations
    if punctuations:
        text = remove_punctuations(text)
    # Alter words with repeated alphabets
    if remove_extra_repeated_alpha:
        text = remove_extra_repeated_alpha(text)
    # remove hashtag
    if hashtag:
        text = remove_hashtag(text)
    #remove latin and number
    if latin:
        text = remove_latin(text)
        
    # normalize text
    if norm:
        text = normalizer.normalize(text)

    # tokenize text
    tokens = word_tokenize(text)
    # remove stop words
    if stopwords:
        tokens = [word for word in tokens if word not in sw]
    # lemmatize words
    if lemmatize:
        tokens = [lemma.lemmatize(word) for word in tokens]
    text = ' '.join(tokens)

    return text
```

```python
9 # Delete row if it is mixed
def delmixed(sentences):
  indexes = []
  for i in range(len(sentences)):

    israisi = 0
    ishemati = 0
    iszakani = 0
    ismehr = 0
    isjalili = 0
    isrezai = 0
    ishashemi = 0

    sentences['text'][i] = sentences['text'][i].replace("ي","ی")
    sentences['text'][i] = sentences['text'][i].replace(b,a)

    tok = sentences['text'][i]
    if 'رییسی' in tok or 'رئیسی' in tok:
      israisi = 1
    if 'همتی' in tok:
      ishemati = 1
    if 'زاکانی' in tok:
      iszakani = 1
    if 'علی زاده' in tok or 'علیزاده' in tok:
      ismehr = 1
    if 'جلیلی' in tok:
      isjalili = 1
    if 'رضایی' in tok or 'رضائی' in tok:
      isrezai:1
    if 'هاشمی' in tok:
      ishashemi = 1

    if israisi + ishemati + iszakani + ismehr + isjalili + isrezai + ishashemi > 1:
      indexes.append(i)

  sentences = sentences.drop(indexes)
  sentences = sentences.reset_index(drop=True)
  return sentences
```

# Remove Short Tweets

```python
9 # Delete row if len token < 3
def remove3(sentences):
  indexes = []
  for i in range(len(sentences)):
    tok = word_tokenize(sentences['text'][i])
    if len(tok) <=3:
      indexes.append(i)

  sentences = sentences.drop(indexes)
  sentences = sentences.reset_index(drop=True)
  return sentences
```
